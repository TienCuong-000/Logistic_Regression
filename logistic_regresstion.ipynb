```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score  
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('D:\\Documents\\california_housing_test.csv')
print("--- Data ---")
print(data.nlargest(5, 'median_house_value'))
x=data.drop('median_house_value', axis=1) # Các đặc trưng đầu vào
y=data['median_house_value']              # Giá trị cần dự đoán
med=np.median(y)
y_bool = (y > med).astype(int)  # Chuyển đổi thành bài toán phân loại
x_train, x_test , y_train , y_test = train_test_split(x,y_bool, train_size=0.8, random_state=42)
```

```python
model= LogisticRegression(max_iter=4000)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# In thêm thông tin
print("\n--- Thông tin mô hình ---")
print("Số mẫu train:", x_train.shape[0])
print("Số mẫu test:", x_test.shape[0])

print("\n--- 10 dự đoán đầu ---")
print("Pred:", y_pred[:10])
print("True:", y_test.values[:10])

from sklearn.metrics import  classification_report


print("\n--- Classification Report ---")
print(classification_report(y_test, y_pred))
```
